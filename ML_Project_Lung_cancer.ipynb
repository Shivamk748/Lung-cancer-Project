# Project:-  Lung Cancer Prediction Based on Lifestyle and Medical Factors


**Project Objective:**

Lung Cancer Prediction Based on Lifestyle and Medical Factors
The objective of this project is to develop a machine learning model to predict the likelihood of lung cancer presence in patients based on a set of lifestyle and medical symptoms. This model aims to assist in early detection and improve diagnostic accuracy by evaluating a patient's risk level through specific indicators, making it a useful tool for healthcare providers.

**Problem Statement:**

Lung cancer remains one of the leading causes of cancer-related deaths worldwide. Early diagnosis is crucial for effective treatment and improving survival rates. This project uses machine learning to analyze patient data based on lifestyle habits and medical symptoms, predicting the presence of lung cancer. Each patient is assessed on several factors, including:


#### 1. Gender

 Type: Categorical (M/F)

Description: Identifies the gender of the patient. Studies have shown that lung cancer risks can vary between males and females due to biological and lifestyle differences.

#### 2. Age
Type: Numerical

Description: Represents the patientâ€™s age. Age is a significant factor, as the risk of lung cancer generally increases with age.

#### 3. Smoking
Type: Binary (YES=2, NO=1)

Description: Indicates whether the patient is a smoker. Smoking is the leading cause of lung cancer, making it one of the most critical risk factors in this analysis.

#### 4. Yellow Fingers  
Type: Binary (YES=2, NO=1)

Description: Signifies whether the patient has yellow-stained fingers, which is often associated with nicotine use. This can be an indicator of heavy or prolonged smoking.

####5. Anxiety
Type: Binary (YES=2, NO=1)

Description: Measures the presence of anxiety in the patient. Anxiety can be both a result of health issues and a contributing factor due to lifestyle choices, indirectly affecting health.

####6. Peer Pressure
Type: Binary (YES=2, NO=1)

Description: Indicates whether peer pressure influences the patient. Peer pressure can affect lifestyle choices, like smoking or drinking, which in turn impacts lung cancer risk.

####7. Chronic Disease
Type: Binary (YES=2, NO=1)

Description: Notes the presence of other chronic diseases. Patients with other chronic conditions may have compromised health, increasing susceptibility to lung cancer.

####8. Fatigue
Type: Binary (YES=2, NO=1)

Description: Indicates if the patient experiences fatigue. Persistent fatigue can be a symptom of lung cancer or other underlying health issues that may correlate with lung cancer risk.

#####9. Allergy
Type: Binary (YES=2, NO=1)

Description: Indicates if the patient has allergies. While not a direct risk factor, allergy symptoms may overlap with lung cancer symptoms, impacting diagnosis.

####10. Wheezing
Type: Binary (YES=2, NO=1)

Description: Refers to the presence of wheezing or breathing difficulties. Wheezing is a respiratory symptom that may be linked to lung issues, including lung cancer.

####11. Alcohol Consumption
Type: Binary (YES=2, NO=1)
Description: Indicates if the patient consumes alcohol. Excessive alcohol use has been linked to a variety of cancers and can be a lifestyle indicator in risk assessment.

#### 12. Coughing
Type: Binary (YES=2, NO=1)

Description: Notes if the patient has a persistent cough. Chronic coughing is a common symptom of lung cancer and thus a significant attribute.

#### 13. Shortness of Breath
Type: Binary (YES=2, NO=1)
Description: Indicates whether the patient experiences shortness of breath. This can be a sign of respiratory issues related to lung cancer.

#### 14. Swallowing Difficulty
Type: Binary (YES=2, NO=1)
Description: Marks any difficulty the patient has swallowing. Although not a direct symptom, it can indicate complications in respiratory or chest health.

#### 15. Chest Pain
Type: Binary (YES=2, NO=1)
Description: Indicates if the patient has chest pain, which can be a severe symptom of lung cancer or other lung-related issues.

#### 16. Lung Cancer
Type: Target Variable (YES/NO)

Description: The target variable indicates the presence (YES) or absence (NO) of lung cancer. The model aims to predict this attribute based on the above factors.


Each of these attributes contributes uniquely to assessing the likelihood of lung cancer, either through direct association with symptoms or through lifestyle factors that influence overall health and lung cancer risk.

By analyzing these attributes, the model seeks to predict lung cancer presence effectively, helping clinicians prioritize high-risk individuals for further testing and intervention.

```python
# Separate features and target
x = df.drop('LUNG_CANCER', axis=1)
y = df['LUNG_CANCER']
```

```python
# Separate categorical and numerical features

categorical = x.select_dtypes(include='object').columns.to_list()

numerical = x.select_dtypes(include=('int64', 'float64')).columns.to_list()
```

```python
# Split into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

### Scaling and Encoding

```python
# Preprocessing: Scaling numerical data and encoding categorical data
scaler = StandardScaler()
encoder = OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False)

# Fit and transform numerical data
x_train_scaled = scaler.fit_transform(x_train[numerical])
x_test_scaled = scaler.transform(x_test[numerical])

# Transform categorical data
x_train_encoded = encoder.fit_transform(x_train[categorical])
x_test_encoded = encoder.transform(x_test[categorical])
```

```python

# Combine the scaled numerical features with encoded categorical features
x_train_final = np.hstack([x_train_scaled, x_train_encoded])
x_test_final = np.hstack([x_test_scaled, x_test_encoded])
```

## 3(b).  Hyper parameter tuning

```python

# Define your parameter grids
param_grids = {
    "Logistic Regression": {'C': [0.1, 1, 5,20,50,100], 'solver': ['lbfgs', 'liblinear']},
    "Support Vector Classifier": {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']},
    "Decision Tree": {'max_depth': [5, 10, 15], 'min_samples_split': [2, 5, 10]},
    "Random Forest": {'n_estimators': [50, 100], 'max_depth': [10, 15], 'min_samples_split': [2, 5]}

}

# Initialize an empty list for results
best_params_results = {}

# Loop through each model and perform Grid Search
for model_name, model_instance in models.items():
    param_grid = param_grids[model_name]

    grid_search = GridSearchCV(model_instance, param_grid, cv=10, scoring='accuracy',n_jobs=-1)
    grid_search.fit(x_train_final, y_train)

    best_params_results[model_name] = grid_search.best_params_
    print(f"Best Hyperparameters for {model_name}:", grid_search.best_params_)

```

# Step 4. Model Evaluation:

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns

# Function to evaluate and display all metrics for each model
def evaluate_model(name, model, x_test_final, y_test):
    # Predict and evaluate
    y_pred = model.predict(x_test_final)
    #y_prob = model.predict_proba(x_test)[:, 1] if hasattr(model, "predict_proba") else None  # For ROC

    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, pos_label='YES')  # Adjust pos_label based on  target variable, 1 taken as encoded previously
    recall = recall_score(y_test, y_pred, pos_label='YES')
    f1 = f1_score(y_test, y_pred, pos_label='YES')

    # Display metrics
    print(f"{name} Evaluation Metrics:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}\n")

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(f"{name} - Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

# Train and evaluate each model
for name, model in models.items():
    model.fit(x_train_final, y_train)
    evaluate_model(name, model, x_test_final, y_test)
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Function to evaluate and plot ROC Curve for each model
def evaluate_model(name, model, x_test_final, y_test):
    # Check if the model has a predict_proba method (not all models do)
    if hasattr(model, "predict_proba"):
        # Get the probabilities of the positive class (class 1)
        y_proba = model.predict_proba(x_test_final)[:, 1]

        # Compute ROC curve and AUC score
        fpr, tpr, _ = roc_curve(y_test, y_proba,pos_label='YES')
        roc_auc = auc(fpr, tpr)

        # Plot the ROC curve
        plt.figure()
        plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f"{name} - ROC Curve")
        plt.legend(loc="lower right")
        plt.show()
    else:
        print(f"{name} does not support probability predictions (predict_proba)")

# Train and evaluate each model
for name, model in models.items():
    model.fit(x_train_final, y_train)
    evaluate_model(name, model, x_test_final, y_test)
```
